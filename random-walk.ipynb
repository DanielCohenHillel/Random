#### ðŸ‘‰ Imports

```python
import numpy as np
import matplotlib.pyplot as plt

plt.style.use("ggplot")
```

#### ðŸ‘‰ Constants

```python
step_num   = 100
res        = 1000
step_width = 1
```

#### ðŸ‘‰ Simulate walks

```python
rand      = (np.random.random([step_num, res]) - 0.5)*2*step_width
walks     = np.cumsum(rand, axis=0)
distances = np.abs(walks)
avg_dist  = np.average(distances, axis=1)
```

#### ðŸ‘‰ Plot average walk graph

```python
plt.title("Avarage Distance of a Random Walk")
plt.xlabel(r'Step Nubmer $N$')
plt.ylabel(r'Distance From the Origin')

plt.plot(avg_dist);
```

#### ðŸ‘‰ Linear fit
We know that the average distance of a random walk is proportional to $\sqrt{N}$ where $N$ is the number of random steps. This means that the average distance squared should be a straight line and we can fit it's parameters

```python
n_arr = np.linspace(1, step_num, step_num)
m, b  = np.polyfit(n_arr, avg_dist**2, 1)

print(f'{m}x + {b}')
```

#### ðŸ‘‰ Plot linear fit

```python
plt.title("Avarage Distance of a Random Walk Squared")
plt.xlabel(r'Step Nubmer $N$')
plt.ylabel(r'Distance From the Origin Squared')

plt.plot(avg_dist**2);
plt.plot(m*n_arr + b, '--');
```

#### ðŸ‘‰ Plot fit on original graph

```python
plt.title("Avarage Distance of a Random Walk")
plt.xlabel(r'Step Nubmer $N$')
plt.ylabel(r'Distance From the Origin')

plt.plot(avg_dist);
plt.plot(np.sqrt(np.abs(m*n_arr + b)), '--');
```
